**Convolutional Neural Networks**
You should build an end-to-end machine learning pipeline using a convolutional neural network model. In particular, you should do the following:

Load the mnist dataset using Pandas. You can find this dataset in the datasets folder.
Split the dataset into training and test sets using Scikit-Learn.
Build an end-to-end machine learning pipeline, including a convolutional neural network model.
Optimize your pipeline by validating your design decisions.
Test the best pipeline on the test set and report various evaluation metrics.
